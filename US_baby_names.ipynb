```python
# Unique names over Time
from collections import defaultdict
```

```python
uniquenames = defaultdict(int)
```

```python
uniquenames
```

```python
for year in range(1880,2021):
    todos = parse_dataset(year)
    uniquenames[year] = todos['name'].nunique()
```

```python
uniqueyearname = pd.DataFrame.from_dict(uniquenames, orient='index', columns=['frequency'])
uniqueyearname
```

```python
plt.plot(uniqueyearname.index, uniqueyearname['frequency'])

plt.xlabel('Time')
plt.ylabel('Unique Names')
plt.title('Unique names over time');
```
